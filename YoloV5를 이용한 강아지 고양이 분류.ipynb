```python
import json
import pandas as pd
import os
import ast
import shutil
import random
```

```python
file_path = '/content/annotation/annotations/instances_val2017.json'
with open(file_path, "r") as json_file:
    data = json.load(json_file)
```

```python
data.keys()
```

```python
data['images']
```

```python
df_info = {}
df_info['filename'] = [filename['file_name'] for filename in data['images']]
df_info['width'] = [filename['width'] for filename in data['images']]
df_info['height'] = [filename['height'] for filename in data['images']]
df_info['id'] = [filename['id'] for filename in data['images']]
```

```python
print(len(df_info['id']))
print(len(df_info['filename']))
print(len(df_info['width']))
print(len(df_info['height']))
```

```python
df = pd.DataFrame(df_info)
```

```python
df.head()
```

```python
info_df = {}
info_df['image_id'] = [file['image_id'] for file in data['annotations']]
info_df['bbox'] = [file['bbox'] for file in data['annotations']]
info_df['class'] = [file['category_id'] for file in data['annotations']]
```

```python
info_df = pd.DataFrame(info_df)
info_df.head()
```

```python
info_df['filename'] = info_df.apply(lambda x: df.loc[df['id'] == x['image_id'], 'filename'].values[0], axis=1)
info_df['width'] = info_df.apply(lambda x: df.loc[df['id'] == x['image_id'], 'width'].values[0], axis=1)
info_df['height'] = info_df.apply(lambda x: df.loc[df['id'] == x['image_id'], 'height'].values[0], axis=1)
```

```python
info_df
```

```python
cat_dog_df = info_df[(info_df['class'] == 18) | (info_df['class'] == 17)]
```

```python
cat_dog_df
```

```python
len(set(cat_dog_df['filename']))
```

```python
file_folder = '/content/val/val2017'
cat_dog_df['file_name'] = cat_dog_df['filename'].apply(lambda x: os.path.join(file_folder, x))
```

```python
cat_dog_df
```

```python
def convert_bbox_to_yolo_revise(bbox, image_width, image_height):
  x_center = (bbox[0] + bbox[2]) / (2 * image_width)
  y_center = (bbox[1] + bbox[3]) / (2 * image_height)
  yolo_width = bbox[2] / image_width
  yolo_height = bbox[3] / image_height
  yolo_bbox = [x_center, y_center, yolo_width, yolo_height]
  return yolo_bbox
```

```python
# cat_dog_df['bbox'] = cat_dog_df['bbox'].apply(lambda box: ast.literal_eval(box))
cat_dog_df['bbox_yolo'] = cat_dog_df.apply(lambda row: convert_bbox_to_yolo_revise(row['bbox'], row['width'], row['height']), axis=1)
```

```python
cat_dog_df
```

```python
cat_dog_df['class'] = cat_dog_df['class'].apply(lambda x: x-17)
```

```python
cat_dog_df.to_csv('left_label.csv', index=False)
```

```python
train_folder = '/content/train/images'
val_folder = '/content/validation/images'
image_folder = '/content/val/val2017'
image_move = '/content/images'
```

```python
if not os.path.exists(train_folder):
    os.makedirs(train_folder)
if not os.path.exists(val_folder):
    os.makedirs(val_folder)
if not os.path.exists(image_move):
    os.makedirs(image_move)
```

```python
image_filenames = [filename for filename in os.listdir(image_folder) if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
len(image_filenames)
```

```python
image_filenames[0]
```

```python
image_usage = cat_dog_df['filename'].tolist()
for image in image_usage:
  if image in image_filenames:
    departure = os.path.join(image_folder, image)
    destination = os.path.join(image_move, image)

    shutil.copy(departure, destination)
```

```python
print(len(os.listdir('images')))
```

```python
images = os.listdir('images')
random.shuffle(images)

train_ratio = 0.8
total_images = len(images)
train_count = int(train_ratio * total_images)
```

```python
train_count
```

```python
for i, image_filename in enumerate(images):
    source_path = os.path.join('images', image_filename)
    if i < train_count:
        destination_path = os.path.join(train_folder, image_filename)
    else:
        destination_path = os.path.join(val_folder, image_filename)
    shutil.copy(source_path, destination_path)
    print(f"Copied {image_filename} to {train_folder if i < train_count else val_folder}")
```

```python
print(len(os.listdir('train/images')))
print(len(os.listdir('validation/images')))
```

```python
train = os.listdir(train_folder)
val = os.listdir(val_folder)

train_data = cat_dog_df[cat_dog_df['filename'].isin(train)]
val_data = cat_dog_df[cat_dog_df['filename'].isin(val)]
```

```python
def create_yolo_label_txt(row):
    class_id = row['class']
    bbox = row['bbox_yolo']
    yolo_label = f"{class_id} {bbox[0]:.6f} {bbox[1]:.6f} {bbox[2]:.6f} {bbox[3]:.6f}"
    return yolo_label
```

```python
# 'train_labels' 폴더 생성
train_labels_folder = '/content/train/labels'
if not os.path.exists(train_labels_folder):
    os.makedirs(train_labels_folder)

# YOLO 레이블 생성 및 .txt 파일 생성
for index, row in train_data.iterrows():
    txt_filename = row['filename'].replace('.jpg', '.txt')
    txt_filepath = os.path.join(train_labels_folder, txt_filename)

    yolo_label = create_yolo_label_txt(row)

    with open(txt_filepath, 'w') as f:
        f.write(yolo_label)

print("YOLO label .txt files created in 'train_labels' folder.")
```

```python
# 'train_labels' 폴더 생성
train_labels_folder = '/content/validation/labels'
if not os.path.exists(train_labels_folder):
    os.makedirs(train_labels_folder)

# YOLO 레이블 생성 및 .txt 파일 생성
for index, row in val_data.iterrows():
    txt_filename = row['filename'].replace('.jpg', '.txt')
    txt_filepath = os.path.join(train_labels_folder, txt_filename)

    yolo_label = create_yolo_label_txt(row)

    with open(txt_filepath, 'w') as f:
        f.write(yolo_label)

print("YOLO label .txt files created in 'train_labels' folder.")
```
